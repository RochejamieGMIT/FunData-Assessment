Find out why the log of 0 is undefined? 

Formula for Log:

 $Log_a(b) = x$
 
 $a = b^x$


Example Formula:

 $Log_a(0) = x$
 
 $a=0^x$
 
 $a = 0$

Log(0):

$log_{10}0  = x$

$10 = 0^x$

$10 \neq 0$

$\therefore Log(0)$ is undefined

When log is converted to a = b^x, any non-zero base will solve to = 0^x, this can not be true as a a can not equal 0^x as this is 0. Which is untrue. Therefore log(0) is undefined.
