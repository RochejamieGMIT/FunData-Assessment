The formula for 4 tails is 10!/(6!4!) = 210, interestingingly though, it can be calculated in terms of heads and return the same answer. 

if you want 4 tails, It means you must have 6 heads.
6 heads = 10!/(10-6)!*6! = 10!/(4!*6!) = 210
4 tails = 10!/(10-4)!*4! = 10!/(6!*4!) = 210

1. Dunning-Kruger Effect 

People with limited knowledge and perhaps lack of skill often, over-estimate their abilities and knowledge when it comes to topics they may not be well versed in, compared to capable people who may lack confidence in their knowledge in the same subject despite being more familiar with the subject. 

2. The Hindsight Bias

It is easier to see faults or assume an outcome was easy to predict when you know the outcome. People can spot decisions that might have led to a bad outcome after the fact, and deem them obvious, when in fact as the decisions and outcomes were unfolding the negative affects the decisions might have had were not so apparent. 

Sports fans often criticise coaches and managers when making changes to the team after a loss, even though at the time the changes did not appear to be detrimental 

3. Anchoring Bias

When receiving information, the first piece of information carries a lot of weight, and can influence a person’s decisions more than they might expect. Companies often have a product that is purposely overpriced, this give consumers a false idea of the cost of products, allowing companies to mark up the cost of ordinary products, as even though they are over charging it appears to be a good deal in comparison to the overpriced item.

# Outliers Exercise 3

Adapt the above code and plots so that the overall plot is inversely propertional and the individual groups are directly proportional.



```python
# Create a series of x vectors.
xs = np.array([np.linspace(i, i + 2.0, 20) for i in range(10)])
```

```python
# Create a corresponding series of y vectors.
ys = np.array([1.0 * -xs[i] + 2 * i for i in range(len(xs))])

# Add some noise.
ys = -ys - np.random.normal(0.0, 1.0, ys.shape)
```

```python
# Create global list of x and y values.
x = xs.flatten()
y = ys.flatten()
```

```python
# Plot.
plt.plot(x, y, '.');
```

```python
# Best fit line slope and y intercept.
m, c = np.polyfit(x, y, 1)
```

```python
# Create a steadily increasing x range for plotting.
xbestfit = np.linspace(x.min(), x.max(), 1000)
# Create corresponding y values form best fit line.
ybestfit = m * xbestfit + c
```

```python
# Plot points.
plt.plot(x, y, '.')

# Plot line.
plt.plot(xbestfit, ybestfit);
```

```python
# Plot the same points, but now with a different colour for each range.
for i in range(len(xs)):
    plt.plot(xs[i], ys[i], '.')
    
```

```python
# Same plot but with best fit lines for each category.
for i in range(len(xs)):
    # Plot this category.
    plt.plot(xs[i], ys[i], '.')
    # Fit line to category.
    m, c = np.polyfit(xs[i], ys[i], 1)
    # Create steadily increasing x range for plot.
    xbestfit = np.linspace(xs[i].min(), xs[i].max(), 50)
    # Corresponding y values based on best fit line.
    ybestfit = m * xbestfit + c
    # Plot best fit line for category.
    plt.plot(xbestfit, ybestfit, color='lightgrey');
```
