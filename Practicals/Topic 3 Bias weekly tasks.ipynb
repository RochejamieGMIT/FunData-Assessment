1. Dunning-Kruger Effect 

People with limited knowledge and perhaps lack of skill often, over-estimate their abilities and knowledge when it comes to topics they may not be well versed in, compared to capable people who may lack confidence in their knowledge in the same subject despite being more familiar with the subject. 

2. The Hindsight Bias

It is easier to see faults or assume an outcome was easy to predict when you know the outcome. People can spot decisions that might have led to a bad outcome after the fact, and deem them obvious, when in fact as the decisions and outcomes were unfolding the negative affects the decisions might have had were not so apparent. 

Sports fans often criticise coaches and managers when making changes to the team after a loss, even though at the time the changes did not appear to be detrimental 

3. Anchoring Bias

When receiving information, the first piece of information carries a lot of weight, and can influence a person’s decisions more than they might expect. Companies often have a product that is purposely overpriced, this give consumers a false idea of the cost of products, allowing companies to mark up the cost of ordinary products, as even though they are over charging it appears to be a good deal in comparison to the overpriced item. 

Some companies have for instance, hoodies priced over 250€, This allows them to charge 100-150€ for regular hoodies and have it appeared as a good deal. Even though the hoodie might only be worth 70€.

# Bias Exercise 2

Show that the difference between the standard deviation calculations is greatest for small sample sizes.

For smaller sample sizes, the standard deviation is more greatly affected by outliers for smaller sample sizes, less data points offer less of a chance to average the difference from the mean. 

In the snippet of code seen below, taken from the Bias note book the mean of the numbers is the same, but as the first set of numbers has 1 outlier that is quite far from the mean, the standard deviation is quite large. 

Larger data sets can have more values to minimise this harsh effect of outliers or show the true properties of the data.

```python
import numpy as np
import matplotlib.pyplot as plt

# A list of numbers - four small and one big.
numbers1 = np.array([1, 1, 1, 1, 10])

# Their mean.
np.mean(numbers1) #mean is 2.8
std1 = np.std(numbers1)

# A list of numbers - all close to each other.
numbers2 = np.array([2, 2, 3, 3, 4])

# Their mean.
np.mean(numbers2) # mean is 2.8
std2 = np.std(numbers2)

```

```python
std1 # Larger std as the outlier heavily affects the outcome as the data set is too small to off set the affects of outliers
```
